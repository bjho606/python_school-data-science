```python
import pandas as pd

xl_file = '/Users/JaehoByun/JB/_School/2021_2 데이터사이언스/과제및시험/db_score.xlsx'
df = pd.read_excel(xl_file)
# print(df)
```

```python
import numpy as np

df_midterm = df['midterm']
df_final = df['final']
df_score = df['score']

midterm_mean = np.mean(df_midterm)
midterm_median = np.median(df_midterm)
final_mean = np.mean(df_final)
final_median = np.median(df_final)
score_mean = np.mean(df_score)
score_median = np.median(df_score)
print(midterm_mean, final_mean, score_mean)
print(midterm_median, final_median, score_median)

from scipy import stats
mode_grade = stats.mode(df['grade'])
print(mode_grade[0])

midterm_var = np.var(df_midterm)
midterm_stdev = np.std(df_midterm)
final_var = np.var(df_final)
final_stdev = np.std(df_final)
score_var = np.var(df_score)
score_stdev = np.std(df_score)
print(midterm_var, final_var, score_var)
print(midterm_stdev, final_stdev, score_stdev)
```

```python
import matplotlib.pyplot as plt

p = np.linspace(0, 100, 20)
y1 = np.percentile(df_midterm, p, interpolation='linear')
y2 = np.percentile(df_final, p, interpolation='linear')
y3 = np.percentile(df_score, p, interpolation='linear')

# plt.plot(p, y1)
# plt.title('Midterm Percentile')
# plt.show()

# plt.plot(p, y2)
# plt.title('Final Percentile')
# plt.show()

# plt.plot(p, y3)
# plt.title('Score Percentile')
# plt.show()

# plt.subplot(1,3,1)
# plt.plot(p, y1)
# plt.title('Midterm Percentile')

# plt.subplot(1,3,2)
# plt.plot(p, y2)
# plt.title('Final Percentile')

# plt.subplot(1,3,3)
# plt.plot(p, y3)
# plt.title('Score Percentile')

# plt.show()

fig, ax = plt.subplots(1, 3)
fig.suptitle('< Percentile >')
fig.set_size_inches(10, 3)
ax[0].plot(p, y1)
ax[0].set_title("Midterm Percentile")

ax[1].plot(p, y2)
ax[1].set_title("Final Percentile")

ax[2].plot(p, y3)
ax[2].set_title("Score Percentile")

plt.subplots_adjust(wspace=0.5)
fig.tight_layout()
plt.show()
```
